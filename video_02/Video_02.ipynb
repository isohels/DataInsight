# This video we will grab the covid cases data from the api using webscrapping in python.

 - The data is taken from the api ('https://api.covid19india.org/states_daily.json'), using urllib.request
 - Then data is converted into json format 


```python
import urllib.request
import json
import requests
import html5lib
import csv
```

<h1>Hitting the api</h1>

```python
url = 'https://api.covid19india.org/states_daily.json'
```

<h1>Parsing the file into json format</h1>

```python
json_obj = urllib.request.urlopen(url)
print(json_obj)
```

<h1>As the states names are shortend we will create another list that will replace the states column inside final dataframe. </h1>

<h1>New List</h1>

```python
stateNames = ["Andaman and Nicobar Islands",
"Andhra Pradesh",
"Arunachal Pradesh",
"Assam",
"Bihar",
"Chandigarh",
"Chhattisgarh",
"Daman and Diu",
"Delhi",
"Dadra and Nagar Haveli",
"Goa",
"Gujarat",
"Himachal Pradesh",
"Haryana",
"Jharkhand",
"Jammu and Kashmir",
"Karnataka",
"Kerala",
"Ladakh",
"Lakshadweep",
"Maharashtra",
"Meghalaya",
"Manipur",
"Madhya Pradesh",
"Mizoram",
"Nagaland",
"Odisha",
"Punjab",
"Puducherry",
"Rajasthan",
"Sikkim",
"Telangana",
"Tamil Nadu",
"Tripura",
"total",
"un",
"Uttar Pradesh",
"Uttarakhand",
"West Bengal"]

len(stateNames)
# The List Sequence needs to be same as we are going to replace json file's state name by our stateNames list
```

<h1>List from api</h1>

```python
states = ["an","ap","ar","as","br","ch","ct","dd","dl","dn","ga","gj","hp","hr","jh","jk","ka","kl","la","ld","mh","ml","mn","mp","mz","nl","or","pb","py","rj","sk","tg","tn","tr","tt","un","up","ut","wb"]
```

```python


json_obj =urllib.request.urlopen(url)
data = json.load(json_obj)
csv_file = open('covid_india.csv','w')
csv_writer =  csv.writer(csv_file)
csv_writer.writerow(['states','cases','date'])


for item in data['states_daily']:
    dates = (item['date'])
    print()
    for i in states:
        cases = (item[i])
        # print(cases)
        csv_writer.writerow([i,cases,dates])
    
csv_file.close()
```
